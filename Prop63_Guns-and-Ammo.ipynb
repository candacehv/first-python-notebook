By Candace Holcombe-Volke
Prop 63 was a proposition on the ballot in California during the November 2016 elections. A supporting vote for this proposition indicates a favorable opinion of a ban on possesion and carrying of large scale magazines and a requirement that certain individuals have a background check prior to authorization to purchase ammunition. A no vote indicates the opposite opinion.

Import pandas

```python
import pandas as pd
```

Read in the Cal-Access data 

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head()
```

```python
props.info()
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contribs.head()
```

Organize table using reset_index()

```python
props.prop_name.value_counts().reset_index()
```

Find number of entries for each proposition using value_counts

```python
props.prop_name.value_counts()
```

create variable and show results in a DataFrame using []

```python
props[props.prop_name == "PROPOSITION 063- FIREARMS. AMMUNITION SALES. INTIATIVE STATUTE."]
```

```python
prop63 = props[props.prop_name == "PROPOSITION 063- FIREARMS. AMMUNITION SALES. INTIATIVE STATUTE."]
```

```python
prop63.head()
```

```python
prop63.info()
```

Merged two tables based on the calaccess_committee_id field also using keyword

```python
merged = pd.merge(prop63, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

Sum of Amount column in merged 

```python
merged.amount.sum()
```

```python
merged.info()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

## Total contributions to support measure

```python
support.amount.sum()
```

## Total Contributions to oppose measure

```python
oppose.amount.sum()
```

Percent of the total contributions made in support of Prop 63

```python
support.amount.sum() / merged.amount.sum()
```

Changing default sort from ascending to descending using sort_values()

```python
merged.sort_values("amount", ascending=False)
```

```python
oppose.sort_values("amount", ascending=False).info()
```

Summing amounts using groupby to aggregate contributions within committee_name_x in descending order

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
